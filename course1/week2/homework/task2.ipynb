```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import linalg
```

```python
def func(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
xlist = np.arange(1, 15, 0.01)
ylist = [func(x) for x in xlist]
```

```python
fig = plt.plot(xlist, ylist)
```

```python
def getA(x):
    l = []
    for i in x:
       for j in range(len(x)):
            l.append(i**j)
    return np.array(l).reshape(len(x), len(x))

def getB(x):
    l = []
    for i in x:
        l.append(func(i))
    return np.array(l)

def genF(w,x):
    f=0
    for i in range(len(w)):
        f=f+w[i]*x**i
    return f
```

```python
slau = ((1, 15), (1, 8, 15), (1, 4, 10, 15))
for tp in slau:
    print(tp)
    A = getA(tp)
    B = getB(tp)
    w = linalg.solve(A, B)
    slau_ylist = [genF(w,x) for x in xlist]
    fig = plt.plot(xlist, ylist, xlist, slau_ylist)
    print(w)
```
